1. Importing Data and Libraries

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.offline as py
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np
````

````python
# Changing each variable name from attribute to its name
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
data = pd.read_csv(url, names=names)
print(data.head())
````

````python
# Bar plot for categorical data
plt.figure(figsize=(12, 6))
sns.countplot(x='income', data=data, order = data['income'].value_counts().index)
plt.title('Frequency of Each Income Category')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
````

````python
# Box-plot for correlation between race and education

plt.figure(figsize=(12, 6))
sns.boxplot(x='race', y='education', data=data)
plt.title('Box Plot of Race and Education')
plt.xlabel('Race')
plt.ylabel('Education')
plt.show()
````

2. The second hypothesis is that there is a correlation between income and age.

````python
# Box-plot for correlation between income and age

plt.figure(figsize=(12, 6))
sns.boxplot(x='income', y='age', data=data)
plt.title('Box Plot of income and age')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
````

Conclusion: By looking at the box plot, those who make <=50k, the average age is in the mid 30s. On the other hand, those who make >50k, the average age is in the mid 40s. We can draw the conclusion that people with more age tend to have higher income. [link text](https://)
