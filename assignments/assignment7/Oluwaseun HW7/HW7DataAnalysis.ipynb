## 3. Hypothesis Formulation

1. I hypothesize that there is a positive correlation between chlorides levels and the pH of the alcohol.

2. I hypothesize that citric acid has the greatest correlation to alcohol quality than any other features of the alcohol listed.

## 4. Hypothesis Testing

```python
print('1. I hypothesize that there is a positive correlation between chlorides levels and the pH of the alcohol.')


# Scatter plot with regression line
sb.regplot(x='chlorides', y='pH', data=df)

# Add labels and title
plt.xlabel('Chlorides Levels')
plt.ylabel('pH of Alcohol')
plt.title('Correlation Between Chlorides Levels and pH')

# Display the plot
plt.show()
```

#### The graph is a regression plot graph. The dots lying to the left show that most red wines vary in pH from 3.0 to 3.6, with some outliers between 2.8 and 3.0 and 3.6 to 4.0. On the left as well, most wines are around 0.1 for their chloride level. However, as chloride levels increase, the pH for all wines gradually decrease. The regression line falls down as we move to the right of the graph, indiciating a negative correlation between the two. This does not support my hypothesis.

```python
print('2. I hypothesize that citric acid has the greatest correlation to alcohol quality than any other features of the alcohol listed.')

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title('Correlation Heatmap Features of Alcohol')

# Display the plot
plt.show()
```

#### This graph is a correlation heatmap that finds the correlation between each feature and associates it with a value between -1 and 1. The colors from blue to red represent the intensity of the correlation with a dard red being 1 (positive/absolute association) and dark blue being -1 (negative/absolute association). My hypothesis is not supported by this heatmap since the quality columns show that alcohol level having the greatest positive correlation number (0.48) but citric acid falls behind with a correlation number of 0.23.
