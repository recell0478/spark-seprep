#### Look into the counts for each gender

```python
gender_count = df.filter(['sex']).value_counts().reset_index()
gender_count.columns = ['Men', 'Women']
plt.bar(gender_count['Men'], gender_count['Women'])
plt.title('Number of men and women')
plt.show()
```

#### Look into the score counts

```python
# find the counts of each score
scores = df.filter(['G3']).value_counts(sort=False).reset_index()
scores.columns = ['Score', 'Counts']
plt.bar(scores['Score'], scores['Counts'])
plt.title('Counts of all scores')
plt.show()
```

#### Check the counts for each age group 

```python
age_counts = df.filter(['age']).value_counts().reset_index()
age_counts.columns = ['Age', 'Count']
plt.bar(age_counts['Age'], age_counts['Count'])
plt.title('Number of students per age group')
plt.show()
```

#### Looking into any correlations between age and final score

```python
# compare age with final score
age_score = df.filter(['age', 'G3'])
age_score.head(15)
```

```python
# graph the average score with age
age_score = df.groupby(['age'])['G3'].mean().reset_index()
age_score.columns = ['Age', 'Average Score']
plt.bar(age_score['Age'], age_score['Average Score'])
plt.title('Average score per age group')
plt.show()
```

# 3. Hypothesis Formulation

#### The first hypothesis is that there is a correlation between age and the final score

```python
weighted_agescore = df.filter(['age', 'G3'])
plt.figure(figsize=(20, 10))
sns.boxplot(x='age', y='G3',data=weighted_agescore)
plt.show()
```

To me it doesn't seem like there is a correlation between a students age and their score. The averages for students aged 15 to 19 seem to be around the same at 11 while the averages for students older than 20 are higher. 

However, there might be some issues with there only being a couple of people older than 20 while there are over 104 people who are 16. When I previously counted the number of students per age group I found that for students aged from 15 to 18, there were around the same number of students in each group. Hwever, there were only 24 students aged 19, 5 students aged 20, 1 student aged 21, and 1 student aged 22. 

If we were to disregard the values for the students older than 19 then we can see that there is no correlation for scores. This makes sense because there really shouldn't be any difference in scores depending on their age since some students can be extremely smart at a young age while some students might not take the course till later in their school time. 

#### The second hypothesis is that there is a correlation between the number of failures and the final score.

```python
#check if there is a correlation between number of failures and average score
failure_score = df.groupby(['failures'])['G3'].mean().reset_index()
failure_score
```

```python
# graph the average score with age and see if there is a correlation
failure_score.columns = ['Failures', 'Average Score']
plt.title('Average score per number of failures')
plt.bar(failure_score['Failures'], failure_score['Average Score'])
plt.xticks(range(0, 4))
plt.show()
```

It seems like there is a correlation between the number of times a student has failed the class and the final score they recieve. We can see that students who didn't fail the class had on average a score around 11. We can see that there is a drop off from never failing to failing the class once where the average score goes from around 11 to around 8. This drop off is less noticiable as students failed the class 2 or 3 times but there is still a decline in the average score.

This makes sense since students that failed the class once obviously are already struggling which explains why the average score is lower. Students who failed the class multiple times are struggling even more and therefore the average score is lower. However, we cant say much about people who haven't failed a class before because the graph only shows scores based on previous failures. Theres a chance a student who hasn't failed before could get a higher score or lower score.
