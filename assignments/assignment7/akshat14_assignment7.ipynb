# 1. First hypothesis is there is a corelation between level of education and income

```python
data['income_binary'] = (data['income'] == '>50K').astype(int)
# Calculate the proportion of individuals earning more than 50K by education level
income_education = data.groupby('education').income_binary.mean().sort_values()

# Strip trailing spaces
data['income'] = data['income'].str.strip()
data['education'] = data['education'].str.strip()

# Create a binary income variable
data['income_binary'] = (data['income'] == '>50K').astype(int)

# Group by 'education' and calculate the mean of 'income_binary'
education_income = data.groupby('education')['income_binary'].mean().reset_index()

# Sort the results for better visualization
education_income = education_income.sort_values(by='income_binary', ascending=False)
# Create a bar plot
plt.figure(figsize=(10, 8))
sns.barplot(x='income_binary', y='education', data=education_income, palette='coolwarm')
plt.title('Proportion of Individuals Earning >$50K by Education Level')
plt.xlabel('Proportion Earning >$50K')
plt.ylabel('Education Level')
plt.show()
```

As we can see from the graph the higher the education more likely a person will earn above 50k$. Therefore my hypothesis is true.

# 2. Second hypothesis is that there is a corelation between race and income

```python
# Assuming data is already loaded into the DataFrame named 'data'
# Strip any trailing spaces
data['income'] = data['income'].str.strip()
data['race'] = data['race'].str.strip()

# Create a binary income variable where 1 indicates incomes >50K and 0 otherwise
data['income_binary'] = (data['income'] == '>50K').astype(int)
# Group by 'race' and calculate the mean of 'income_binary'
race_income = data.groupby('race')['income_binary'].mean().reset_index()

# Sort the results for better visualization
race_income = race_income.sort_values(by='income_binary', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='income_binary', y='race', data=race_income, palette='viridis')
plt.title('Proportion of Individuals Earning >$50K by Race')
plt.xlabel('Proportion Earning >$50K')
plt.ylabel('Race')
plt.show()
```

As we can see there is a direct corelation between race and income where certain race earn higher than other races. Therefore my hypothesis is true.
